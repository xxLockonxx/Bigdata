```python
import pandas as pd
```

```python
#CSV파일을 데이터프레임으로 생성
df_exam = pd.read_csv('./file/exam.csv')
df_exam
```

```python
#기본요약통계
df_exam.describe()
```

```python
#데이터 상위,하위 5개 미리보기
df_exam.head()
df_exam.tail()
```

```python
#수학점수 60점 이상
df_exam[df_exam.math >= 60]
```

```python
#수학점수 60점 이상 70점 미만
df_exam[(df_exam.math >= 60) & (df_exam.math < 70)]
```

```python
#수학 오름차순
df_exam.sort_values(by='math')
```

```python
#수학 내림차순
df_exam.sort_values(by='math', ascending=False)
```

```python
#영어점수 80점이상 100이하 이고 내림차순 정렬
df_exam[(df_exam.english >=80) & (df_exam.english <=100)].sort_values(by='english', ascending=False)
```

```python
#파생변수
df_exam['total']=df_exam.math + df_exam.english + df_exam.science
df_exam
```

```python
df_exam['mean']=(df_exam.math + df_exam.english + df_exam.science)/3
df_exam
```

```python
# pandas 데이터프레임 인덱싱
# loc - 컬럼명으로 데이터프레임 인덱싱 (R에서 select함수랑 같은 것)
df1 = df_exam.loc[:,['math','science']]
df1
```

```python
df2 = df_exam.loc[1:4,['math','english']]
df2
```

```python
#iloc - 인덱스 번호로 인덱싱
df3 = df_exam.iloc[1:5, 2:4] #2:4는 시작과 끝
df3
```

```python
df4 = df_exam.iloc[1:5, [2, 3]] #여기선 컬럼 번호
df4
```

```python
df5 = df_exam.iloc[1:5, [2, 4]]
df5
```
